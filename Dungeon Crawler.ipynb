```python
from IPython.display import clear_output
from time import sleep

class Entity:
    def __init__(self, grid, x, y, char="#"):
        self.grid = grid
        self.x = x
        self.y = y
        self.char = char
    
    def update_position(self, char=None):
        if char == None:
            char = self.char
        self.grid.tiles[self.y][self.x] = char

class Actor(Entity):
    def __init__(self, grid, x, y, char="@"):
        super().__init__(self, grid, x, y, char)
        
    def move(self, x, y):
        new_x = self.x + x
        new_y = self.y + y
        if new_x >= 0 and new_x < self.grid.width and new_y >= 0 and new_y < self.grid.height:
            self.update_position(" ")
            self.x = new_x
            self.y = new_y
            self.update_position()
    
    def take_turn(self, action):
        if action == "U":
            self.player.move(0, -1)
        if action == "D":
            self.player.move(0, 1)
        if action == "L":
            self.player..move(-1, 0)
        if action == "R":
            self.player..move(1, 0)
        elif action == "Q":
            self.grid.running = False
        
class Grid:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.player = None
        self.tiles = []
        self.running = True
        self.victory = False
        self.status = "BEGIN YOUR QUEST!"
        
        self.start_game()
        
    def start_game(self):
        self.player = Entity(self, 5, 5, "@")
        self.tiles = self.generate_grid(50, 10)
        self.player.update_position()
        self.main_loop()
        
    def generate_grid(self, width, height):
        return [[" " for x in range(self.width)] for y in range(self.height)]
        
    def display_tiles(self):
        clear_output()
        print("#" * self.width + "##")
        for y in range(self.height):
            print("#", end="")
            for x in range(self.width):
                full_char = self.get_full_char(self.tiles[y][x])
                print(full_char, end="")
            print("#")
        print("#" * self.width + "##")
            
    def alter_tile(self, x, y, character):
        self.tiles[y][x] = character
    
    def display_status(self):
        print(self.status)
    
    def get_full_char(self, char):
        if char == "@":
            return char
        elif char == "^":
            return char
        elif char == "$":
            return char
        else:
            return char

    def main_loop(self):
        while self.running and not self.victory:
            self.display_tiles()
            self.display_status()
            self.status = ""
            sleep(0.5)
            action = input("MAKE YOUR CHOICE: ").upper()
            take_turn(action)
            
```
